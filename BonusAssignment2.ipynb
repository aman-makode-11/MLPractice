```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/sample_data/california_housing_test.csv')
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Handle missing values:
# For example, let's impute missing values in the numerical columns with the median

# Import SimpleImputer from sklearn.impute
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

# Check again for any missing values
print(data.isnull().sum())
```

```python
# Plot a heatmap of the correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Strip leading and trailing spaces from column names
data.columns = data.columns.str.strip()

# Recheck the column names
print(data.columns)
# Check the first few rows to see the correct column names and adjust
print(data.head())
```

```python
# Replace 'MedInc' with the correct column name
# Let's assume 'MedInc' is actually 'median_income'
plt.figure(figsize=(8, 5))
sns.histplot(data['median_income'], kde=True, bins=30)
plt.title('Distribution of Median Income')
plt.show()
```

```python
# Scatter plot between 'median_income' and 'housing_median_age'
plt.figure(figsize=(8, 5))
sns.scatterplot(x='median_income', y='housing_median_age', data=data)  # Changed 'HouseAge' to 'housing_median_age'
plt.title('Relationship between Median Income and House Age')
plt.show()
```

```python
# Export the cleaned data to a new CSV file
data.to_csv('/content/sample_data/cleaned_california_housing.csv', index=False)

# Print the cleaned dataset
print("\nCleaned Dataset:")
print(data.head())  # Display the first few rows of the cleaned data
```
